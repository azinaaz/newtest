```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,3 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y);
```

```python
y1 = np.cos(x)
plt.plot(x, y, 'rx', x, y1, 'bo')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sine and cosine')
plt.legend(['sin', 'cos']);
```

```python
v = np.linspace(-1.4, 1.4, 30)
plt.subplot(2,2,1)
plt.plot(v,v)
plt.subplot(2,2,2)
plt.plot(v,v**2)
plt.subplot(2,2,3)
plt.plot(v,v**3)
plt.subplot(2,2,4)
plt.plot(v,v**4)
plt.show()
```
